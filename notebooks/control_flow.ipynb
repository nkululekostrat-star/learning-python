Problem Statement

You are given a list of temperature readings (in °C) collected hourly from a cold-chain logger:

temperatures = [2.5, 3.1, 4.0, 5.8, 7.2, 6.9, 4.5, 3.0, 1.8, -0.5]

Cold-chain rules:

Safe: 2 ≤ temperature ≤ 6

Warning: temperature < 2 OR 6 < temperature ≤ 8

Critical: temperature > 8 OR temperature < 0

🎯 Your Tasks (in order)

1️⃣ Classify each reading

Using a for loop and if / elif / else, print:

Hour 1: 2.5°C → SAFE

Hour 2: 3.1°C → SAFE
...

Use range() to generate the hour number.

```python
temperatures = [2.5, 3.1, 4.0, 5.8, 7.2, 6.9, 4.5, 3.0, 1.8, -0.5]
hours = list( range(len(temperatures)))

for i in hours:
    if  2 <= temperatures[i] <= 6:
        print("Hours",i+1,":",temperatures[i]," → Safe")
    elif ( 0 < temperatures[i] < 2)  or ( 6 < temperatures[i] <= 8) :
        print("Hours",i+1, ":",temperatures[i], " → Warning")
    else:
        print("Hours",i+1,":", temperatures[i], " → Critical") 
```
