🎯 Exercise: Cold-Chain Classifier Function
Your task

Write a function that:

Takes a list of temperatures

Returns three counts

Stops early if a critical value is detected

Outcome:
Safe: 5
Warning: 4
Critical: 1

f:List[R]→N3

```python
temperatures = [2.5, 3.1, 4.0, 5.8, 7.2, 6.9, 4.5, 3.0, 1.8, -0.5]

def temp_count(t_list):
    safe = 0
    warning = 0
    critical = 0
    for i in t_list:
        if 2 <= i <= 6:
            safe += 1
        elif ( 0 < i < 2)  or ( 6 < i <= 8):    
            warning += 1
        else: 
            critical += 1    
            break
    return  safe, warning , critical

s, w, c = temp_count(temperatures)

print(f"Safe: {s}")
print(f"Warning: {w}")
print(f"Critical: {c}")
```
