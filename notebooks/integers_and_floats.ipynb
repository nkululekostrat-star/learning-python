```python
# integers
x = 100
y = 10.0
print(type(x))
print(type(y))  # Output: <class 'int'>
```

```python
# Summation
summation = x + y
print(summation)  # Output: 22.1

# Subtraction
subtraction = x - y
print(subtraction)  # Output: -0.10000000000000142

# Multiplication
multiplication = x * y
print(multiplication)  # Output: 122.1

# Division
division = x / y
print(division)  # Output: 0.990990990990991

# Modulus
modulus = y % 2
print(modulus)  # Output: 1.1

# Exponentiation
exponentiation = x ** 2
print(exponentiation)  # Output: 121

# Floor Division
floor_division = y // 2
print(floor_division)  # Output: 5.0

# Order of operations
result = (x + y) * (x - y) / x
print(result)  # Output: 0.1999999999999993

# increment and decrement

x = x + 1
y = y + 1

print(f"x is : {x}")
print(f"y is : {y}")

# absolute value
abs_value = abs(y - x)
print(f"abs_value is : {abs_value}")  # Output: 9.0

# round value
round_value = round(y / 3, 1)
print(f"round_value is : {round_value}")  # Output: 3

# comparison
is_equal = (x == y)
print(f"x is_equal to y : {is_equal}")  # Output: False
is_greater = (x > y)
print(f"x is_greater than y : {is_greater}")  # Output: True
is_less = (x < y)
print(f"x is_less than y : {is_less}")  # Output: False
is_not_equal = (x != y)
print(f"x is_not_equal to y : {is_not_equal}")  # Output: True
is_greater_equal = (x >= y)
print(f"x is_greater_equal to y : {is_greater_equal}")  # Output: True
```

```python
# int and float from string
num1 = '100'
num2 = '200.2'
# casting string to int
sum_from_strings = int(num1) + float(num2)
print(f" The sum is : {sum_from_strings}")


```
