```python
# Creating a string variable and printing it
message = "Hello, World!"
print(message)
```

```python
# Using double quotes to include single quotes in the string
message = "This is Bob's world."
print(message)
```

```python
# using single quotes to include double quotes in the string
message = 'She said, "Hello!"'
print(message)
```

```python
# How long is this string?
message = "Hello, World!"
length = len(message)
print(length)
```

```python
# Slicing the string to get the first five characters
message = "Hello, World!"
print(message[6:12])  # Output: World
```

```python
# lower and upper case
message = "Hello, World!"
print(message.lower())  # Output: hello, world!
print(message.upper())  # Output: HELLO, WORLD!
```

```python
# Counting occurrences of a substring
message = "Hello, World! Welcome to the world of Python."
message = message.lower()  # Convert to lower case for case-insensitive count
count = message.count("world")
print(count)  # Output: 1
```

```python
# Replace a substring
message = "Hello, World!"
new_message = message.replace("World", "Python")
print(new_message)
```

```python
# Concatenation of strings
greeting = "Hello"
world = " World."
full_greeting = greeting + ", " + world
# using f-string for better formatting
formatted_greeting = f"{greeting},{world}"
print(formatted_greeting)  # Output: HelloWorld
```

```python
# directories and methods for string objects
message = "Hello, World!"
print(dir(message))
```
